# Drop missing value

```python
df = df[df.ind_empleado.notnull()] # store data with ind_empleado not null into df as new df
```

# Replacing missing value

```python
df_new = df.drop(['tipodom', 'cod_prov', 'ult_fec_cli_1t', 'conyuemp'], axis=1)
```

```python
df.shape
```

```python
df_new.shape
```

```python
values = {'sexo': 'UN', 'indrel_1mes': 1, 'tiprel_1mes': "UN", 'canal_entrada': 'UN',"nomprov":"MADRID", "ind_nomina_ult1" : 0, "ind_nom_pens_ult1":0}
df_new = df_new.fillna(value=values)
```

```python
df_new["renta"] = df_new.groupby("nomprov").transform(lambda x: x.fillna(x.mean())) 
#df_new["renta"] = df_new.groupby("nomprov")['renta'].transform(mean)
```

```python
na2 = pd.DataFrame(df_new.isnull().sum(), columns=['missing'])
na2["%"] = (na2['missing']/df_new.shape[0])*100
na2
```
